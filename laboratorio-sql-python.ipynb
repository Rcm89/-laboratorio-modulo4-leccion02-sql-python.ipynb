En nuestra base de datos vamos a crear 3 tablas, cada una proveniente de uno de los dataframes con los que hemos trabajado

```python
try:
    # Conectar a la base de datos
    connection = psycopg2.connect(
        host="localhost",
        database="laboratorio-SQL-Python",
        user="postgres",
        password="admin",
        port="5432")
    
    cursor = connection.cursor()
    
except OperationalError as e:
    # Manejo de errores comunes
    if e.pgcode == errorcodes.INVALID_PASSWORD:
        print("Contraseña inválida.")
        
    elif e.pgcode == errorcodes.CONNECTION_EXCEPTION:
        print("Error de conexión.")
    else:
        print(f"Ocurrió un error: {e}")

    
```

```python
# Creamos las tablas en SQL

# 1. Tabla de municipios
query_creacion_municipios = '''
    create table if not exists municipios(
        id_municipio VARCHAR(100) primary key, 
        latitud INT,
        longitud INT
        ); '''


cursor.execute(query_creacion_municipios)
connection.commit()

# 2 Tabla de lugares
query_creacion_lugares = '''
    create table if not exists lugares (
        id_local SERIAL primary key,
        id_municipio VARCHAR(100)
            references municipios(id_municipio)
            on update cascade,
        categoria CHAR(50),
        estado CHAR(50),
        distancia INT,
        link VARCHAR(100),
        nombre VARCHAR(100),
        direccion VARCHAR(300),
        latitud DECIMAL,
        longitud DECIMAL
    );'''

cursor.execute(query_creacion_lugares)
connection.commit()

# 3. Tabla de clima
query_creacion_clima = '''
    create table if not exists clima (
        id_clima SERIAL primary key,
        fecha DATE,
        cielo VARCHAR(50),
        temperatura DECIMAL,
        sensacion DECIMAL,
        racha_max DECIMAL,
        precipitacion DECIMAL,
        nieve DECIMAL,
        humedad_relativa DECIMAL,
        prob_nieve DECIMAL,
        prob_tormenta DECIMAL,
        direccion_viento VARCHAR(10),
        velocidad_viento DECIMAL,
        municipio_id VARCHAR(100)
            references municipios(id_municipio)
            on update cascade,
        hora VARCHAR (20)
    );'''

cursor.execute(query_creacion_clima)
connection.commit()


# Ejecutar las consultas para crear las tablas
cursor.execute(query_creacion_municipios)
cursor.execute(query_creacion_clima)
cursor.execute(query_creacion_lugares)

# Confirmar los cambios
connection.commit() 

```

```python
# Recoger datos de temperatura y sensacion termica de Getafe el 2024-08-29
query_2 = """
SELECT  c.temperatura, c.sensacion
FROM clima c
JOIN municipios m ON c.municipio_id = m.id_municipio
WHERE m.id_municipio = %s AND c.fecha = %s;
"""
cursor2.execute(query_2, ('getafe', '2024-08-29'))
resultado_2 = cursor2.fetchall()

for fila in resultado_2:
    print(fila)
```

```python
# Municipio con la temperatura más alta
query3 = '''
    select m.id_municipio, c.temperatura 
    from clima c
    inner join municipios m
        on c.municipio_id = m.id_municipio
    order by c.temperatura desc
    limit 1;
    '''

cursor.execute(query3)
df_temp_mas_alta = pd.DataFrame(cursor.fetchall()).head() 
df_temp_mas_alta
```

```python
#Lugares con el cielo despejado el 29 de Agosto de 2024

query4 = '''
    select l.nombre
    from lugares l
    inner join clima c on l.id_municipio = c.municipio_id
    where c.cielo = 'Cielo despejado' and c.fecha = '2024-08-29';'''

cursor.execute(query4)
df_despejado = pd.DataFrame(cursor.fetchall())
df_despejado
```

```python
#Municipios con más de 3 lugares

query5 = '''
    select m.id_municipio, count(l.id_local)
    from municipios m
    inner join lugares l on
        m.id_municipio = l.id_municipio
        
    group by m.id_municipio
    having count(l.id_local) > 3;
'''

cursor.execute(query5)
df_municipios_lugares = pd.DataFrame(cursor.fetchall())
df_municipios_lugares
```

```python
query_3 = """
SELECT m.nombre_municipio, l.category, COUNT(*) as total_lugares
FROM lugares l
JOIN municipios m ON l.id_municipio = m.id_municipio
WHERE m.nombre_municipio = %s
GROUP BY m.nombre_municipio, l.category;
"""
cursor2.execute(query_3, ('valencia',))
result_3 = cursor2.fetchall()

for row in result_3:
    print(row)
```
